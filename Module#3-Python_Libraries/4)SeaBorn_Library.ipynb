# Seaborn



*   Data Visualization



```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

Note: Seaborn have some built -in data set

```python
#total bill vs tip dataset
tips=sns.load_dataset('tips')
tips.head()
```

```python
sns.set_theme()
```

```python
sns.relplot(data=tips,x='total_bill',y='tip',col='time',hue='smoker',style='smoker',size='size')
```

```python
# load iris dataset
iris=sns.load_dataset('iris')
iris.head()
```

### Scatterd Plot

```python
sns.scatterplot(data=iris,x='sepal_length',y='sepal_width',hue='species',style='species')
```

```python
sns.scatterplot(data=iris,x='sepal_length',y='petal_length',hue='species',style='species')
```

```python
# loading titanic dataset

titanic=sns.load_dataset('titanic')
titanic.head()
```

```python
sns.countplot(data=titanic,x='class')
```

```python
sns.countplot(x='survived',data=titanic)
```

```python
sns.barplot(data=titanic,x='sex',y='survived',hue='class')
```
