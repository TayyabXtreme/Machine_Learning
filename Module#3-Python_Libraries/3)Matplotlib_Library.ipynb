# Matplotlib



*   Useful for making Plots



```python
#importing matplotlib library

import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x=np.linspace(0,10,100)
y=np.sin(x)
z=np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

## Plotting the Data

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,z)
plt.show()
```

```python
# adding title x-axis and y-axis lables
plt.plot(x,y)
plt.xlabel('Angle')
plt.ylabel('Sin(x)')
plt.title('Sine Wave')
plt.show()
```

```python
# parabola
x=np.linspace(-10,10,20)
y=x**2
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Parabola')
plt.show()
```

```python
plt.plot(x,y,'r+')
plt.show()
```

```python
x=np.linspace(-5,5,20)
plt.plot(x,np.sin(x),'g-')
plt.plot(x,np.cos(x),'r--')
plt.show()
```

## BarPlot

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
lagnuages=['C','C++','JavaScript','Python','PHP']
students=[23,17,35,30,15]
ax.bar(lagnuages,students)
plt.xlabel('lagnuages')
plt.ylabel('students')
plt.show()
```

```python
fig1=plt.figure()
ax1=fig1.add_axes([0,0,1,1])
people=[23,17,35,30,15]
lagnuages=['English','Urdu','Chines','English','Kannada']
ax1.pie(people,labels=lagnuages,autopct='%1.1f%%')
plt.show()
```

### Scatter Plot

```python
x=np.linspace(0,10,30)
y=np.sin(x)
z=np.cos(x)
fig2=plt.figure()
ax2=fig2.add_axes([0,0,1,1])
ax2.scatter(x,y,color='g')
ax2.scatter(x,z,color='b')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_title('Scatter Plot')
plt.show()
```

```python
fig3=plt.figure()
ax3=plt.axes(projection='3d')
z=20*np.random.random(100)
x=np.sin(z)
y=np.cos(z)
ax3.scatter3D(x,y,z,c=z,cmap='Greens')
plt.show()
```
