### scikit learn library basic usage

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.linear_model import LinearRegression
```

```python
#load the data
df=sns.load_dataset('tips')
df.head()
```

```python
X=df[['total_bill']]
y=df['tip']
```

```python
# create a regression line plot
sns.regplot(x='total_bill',y='tip',data=df)
```

```python
model=LinearRegression()
```

```python
model.fit(X,y)
```

```python
value_of_predict=[[50]]
model.predict(value_of_predict)
```
