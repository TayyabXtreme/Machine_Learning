```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
import operator

np.random.seed(0)
x=2-3 * np.random.normal(0, 1, 20)
y=x-2*(x**2)+np.random.normal(-3, 3, 20)

x=x[:, np.newaxis]
y=y[:, np.newaxis]

```

```python
degree=2
poly_features=PolynomialFeatures(degree=degree, include_bias=False)
x_poly=poly_features.fit_transform(x)
model=LinearRegression()
model.fit(x_poly, y)
y_poly_pred=model.predict(x_poly)


plt.scatter(x, y, color='blue', s=10, label='Data points')
sorted_axis=operator.itemgetter(0)
sorted_zip=sorted(zip(x,y_poly_pred), key=sorted_axis)
x, y_poly_pred=zip(*sorted_zip)
plt.plot(x, y_poly_pred, color='red', label='Polynomial regression')
plt.xlabel('x')
plt.ylabel('y')
plt.title(f'Polynomial Regression (degree={degree})')
plt.legend()
plt.show()
```
