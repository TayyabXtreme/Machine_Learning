# Label Encoading:
 converting the labels into numeric form

```python
# importing the Dependencies

import pandas as pd
from sklearn.preprocessing import LabelEncoder
import numpy as np
```

```python
dataset=pd.read_csv('/content/breast_cancer_data.csv')
```

```python
dataset.head()
```

```python
dataset.diagnosis.value_counts()
```

```python
label_encoder=LabelEncoder()
```

```python
labels=label_encoder.fit_transform(dataset.diagnosis)
```

```python
print(labels)
```

```python
dataset['target']=labels
```

```python
dataset.head()
```

```python
dataset['target'].value_counts()
```
